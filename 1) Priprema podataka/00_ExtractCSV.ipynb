```python
import pandas as pd
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("asaniczka/top-spotify-songs-in-73-countries-daily-updated")

print("Path to dataset files:", path)
```

```python
import os
print("Fajlovi u datasetu:", os.listdir(path))
csv_path = os.path.join(path, 'universal_top_spotify_songs.csv')
```

```python
df = pd.read_csv(csv_path)
```

```python
df.head()
```

```python
df["snapshot_date"] = pd.to_datetime(df["snapshot_date"])
```

```python
df = df[df["snapshot_date"] <= "2025-03-07"]
```

```python
df = df.reset_index(drop=True)
```

```python
print(df.head())
print(len(df))
```

```python
df.to_csv('universal_top_spotify_songs.csv', index=False)
```
